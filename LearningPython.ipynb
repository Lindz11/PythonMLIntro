```python
# The next thing I want to tryout is formating datetimes and getting them all to be the same farmat no whatever what country it is from
# Formatting it as YYYY-MM-DD 
computer_date = pd.to_datetime(netflix['date_added']) 
netflix['date_added'] = computer_date
netflix.head()
```

```python
# If you look through the data frame you can see that there are some Nan meaning that it is not filled in 
# We want to be able to fill this is by using fillna fuinction

from datetime import datetime as dt 
netflix['date_added'].fillna(dt(2020,1,1), inplace = True)
netflix
```

```python
# This is very useful for applying something  either over a whole column or over multiple columns
# Lamda functions is a function that works on a single row 
# Lambda x: applies whatever u want like a function 
# In this example we just create a function that adds numbers together

the_adder = lambda x: x + 2101010
the_adder(3565)
```

```python
# If you want to apply this same function to a dataframe you use it with either the map method or the apply method
# If its fot a single row then use the map 
# If its for multiple then use the apply 
netflix["genre_count"] = netflix["listed_in"].map(lambda x: len(x.split(",")))
# What the sentence above says is go through the listed in column and give me the how man genres per movie that is seperated by a comma
netflix.head(10)
```

```python
# When you're a data analyst you recieve data that is not group or sorted and for the most part you do not need all of the data
# Because of this you want to mkae sure that you can use things such as groupby, aggregation functions, and sorting values
# For example maybe you want to see how many movies there are in the dataset by country and so you want to group by count and list them in desc order. 

netflix_movies_by_country_count = netflix.groupby("country")["show_id"].count()
netflix_movies_by_country_count
# You can see from the list that it is grouped and counted but it looks weird because the output is displayed as a group by object 
```

```python
# To get this better formatted you can just use reset_index function 
# This will put it back as a dataframe for you 
# And to output it into descending order you can just do ascending is false
netflix_movies_by_country_count = netflix.groupby("country")["show_id"].count().sort_values(ascending = False).reset_index()
netflix_movies_by_country_count
```

```python
# Next big thing for a data analyst is joining and union 
# A union is taking two datasets and stacking them on top of each toher and a join is doing the same thing but horizontally. 
# Its called union in sql but its called concat in python 

# For an exmaple we can do this with another netflix dataset 

second_netflix_dataset = pd.read_csv('netflix_titles_second.csv')
new_dataset = pd.concat([netflix, second_netflix_dataset])
new_dataset

# It worked as you can see the dataset went from 7000 entries to 8000+ entries.
```

```python
# For join it is dependent on how you want to join, inner join, left join,right join, and full join 
# For example we can join our netflix country show count dataset with the second netflis dataset and see what we get

netflix_merged_datset = pd.merge(left = new_dataset, right = netflix_movies_by_country_count, how ="inner", left_on ="country", right_on ="country")
netflix_merged_datset
```
